-- Пример скачивания файла.

```python
import time

def download_file(file_name):
    print(f"Starting download: {file_name}")
    time.sleep(2)  
    print(f"Finished download: {file_name}")

start_time = time.time()

download_file("file1.zip")
download_file("file2.zip")
download_file("file3.zip")

end_time = time.time()

print(f"Total time taken: {end_time - start_time:.2f} seconds")
```

```python
import time
import threading

def download_file(file_name):
    print(f"Starting download: {file_name}")
    time.sleep(2)  # Simulate network delay
    print(f"Finished download: {file_name}")

start_time = time.time()

threads = []
for each in ['file1.zip', 'file2.zip', 'file3.zip']:
    t = threading.Thread(target=download_file, args=(each,))
    t.start()
    threads.append(t)

for t in threads:
    t.join()

end_time = time.time()

print(f"Total time taken: {end_time - start_time:.2f} seconds")
```

```python
import threading
import time

def show_thread_info():
    print(f"Running in: {threading.current_thread().name}")
    time.sleep(1)

threads = []
for i in range(3):
    t = threading.Thread(target=show_thread_info, name=f"MyThread-{i}")
    t.start()
    threads.append(t)

for t in threads:
    t.join()

print("Main thread finished.")
```

Locks and Race Conditions

```python
import threading
import time

balance = 1000  # shared resource

def withdraw(amount):
    global balance
    # Check then modify (multi-step operation)
    if balance >= amount:
        time.sleep(0.01)  # simulate delay
        balance -= amount
        print(f"Withdrawn {amount}, new balance: {balance}")
    else:
        print("Not enough balance!")

threads = []
for _ in range(5):
    t = threading.Thread(target=withdraw, args=(300,))
    threads.append(t)
    t.start()

for t in threads:
    t.join()

print("Final balance:", balance)
```

I/O TASKS

```python
import time
import requests

urls = [
    "https://www.python.org",
    "https://www.github.com",
    "https://www.wikipedia.org",
    "https://www.stackoverflow.com",
    "https://www.realpython.com"
]

def fetch(url):
    print(f"Fetching {url}")
    resp = requests.get(url)
    print(f"Done {url}, length: {len(resp.text)}")
    return len(resp.text)

start = time.time()
results = []
for url in urls:
    results.append(fetch(url))
end = time.time()

print("Sequential results:", results)
print(f"Sequential time taken: {end-start:.2f} seconds")
```

```python
import time
import requests
from concurrent.futures import ThreadPoolExecutor

urls = [
    "https://www.python.org",
    "https://www.github.com",
    "https://www.wikipedia.org",
    "https://www.stackoverflow.com",
    "https://www.realpython.com"
]

def fetch(url):
    print(f"Fetching {url}")
    resp = requests.get(url)
    print(f"Done {url}, length: {len(resp.text)}")
    return len(resp.text)

start = time.time()
with ThreadPoolExecutor(max_workers=5) as executor:
    results = list(executor.map(fetch, urls))
end = time.time()

print("Threaded results:", results)
print(f"Threaded time taken: {end-start:.2f} seconds")
```

CPU BOUNT TASK

demonstrate with multithreading

```python
import time 

def cpu_bound_task():
    count = 0
    for i in range(10**7):
        count += i
    
start_time = time.time()

for _ in range(4):
    cpu_bound_task()

end_time = time.time()

print(f"Время выполнение программы заняло: {end_time - start_time:.2f}")
```

```python
import time 
from multiprocessing import Process

def cpu_bound_task():
    count = 0
    for i in range(10**7):
        count += i
    
start_time = time.time()

processes = [Process(target=cpu_bound_task) for _ in range(4)]

for each in processes:
    each.start()

for each in processes:
    each.join()

end_time = time.time()

print(f"Время выполнение программы заняло: {end_time - start_time:.2f}")
```

```python

from PIL import Image, ImageFilter
import glob
from concurrent.futures import ProcessPoolExecutor
import time


img_names = glob.glob('images/*.jpg')

def process_image(img_file):
    img = Image.open(img_file)
    img = img.filter(ImageFilter.GaussianBlur(15))
    img.thumbnail((1200,1200))

    filename = img_file.split('\\')[-1]
    img.save(f'processed/{filename}')
    print(f'{filename} - was processed')

if __name__ == "__main__":
    t1 = time.time()

    with ProcessPoolExecutor() as executor:
        executor.map(process_image, img_names)
        
    t2 = time.time()

    print(f"end time: {t2 - t1:.2f}")
```
